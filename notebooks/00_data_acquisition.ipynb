## 5Ô∏è‚É£ –î–µ—Ç–∞–ª—å–Ω—ã–π –∞–Ω–∞–ª–∏–∑ –∫–∞–∂–¥–æ–π –∫–æ–ª–æ–Ω–∫–∏

```python
# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –∞–Ω–∞–ª–∏–∑–∞ –∫–æ–ª–æ–Ω–∫–∏
def analyze_column(df, column):
    print(f'\n{'='*60}')
    print(f'üìä –ê–Ω–∞–ª–∏–∑ –∫–æ–ª–æ–Ω–∫–∏: {column}')
    print('='*60)
    
    # –¢–∏–ø –¥–∞–Ω–Ω—ã—Ö
    print(f'–¢–∏–ø –¥–∞–Ω–Ω—ã—Ö: {df[column].dtype}')
    
    # –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π
    print(f'–£–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π: {df[column].nunique():,}')
    
    # –ü—Ä–æ–ø—É—â–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è
    missing = df[column].isnull().sum()
    print(f'–ü—Ä–æ–ø—É—â–µ–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π: {missing} ({missing/len(df)*100:.2f}%)')
    
    # –î–ª—è —á–∏—Å–ª–æ–≤—ã—Ö –∫–æ–ª–æ–Ω–æ–∫
    if pd.api.types.is_numeric_dtype(df[column]):
        print(f'\n–°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞:')
        print(f'  –ú–∏–Ω–∏–º—É–º: {df[column].min():,.0f}')
        print(f'  25%: {df[column].quantile(0.25):,.0f}')
        print(f'  –ú–µ–¥–∏–∞–Ω–∞: {df[column].median():,.0f}')
        print(f'  –°—Ä–µ–¥–Ω–µ–µ: {df[column].mean():,.0f}')
        print(f'  75%: {df[column].quantile(0.75):,.0f}')
        print(f'  –ú–∞–∫—Å–∏–º—É–º: {df[column].max():,.0f}')
        print(f'  –°—Ç–¥. –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ: {df[column].std():,.0f}')
    
    # –î–ª—è –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö –∫–æ–ª–æ–Ω–æ–∫
    else:
        print(f'\n–¢–æ–ø-10 –∑–Ω–∞—á–µ–Ω–∏–π:')
        top_values = df[column].value_counts().head(10)
        for val, count in top_values.items():
            print(f'  {val}: {count:,} ({count/len(df)*100:.1f}%)')

# –ê–Ω–∞–ª–∏–∑–∏—Ä—É–µ–º –≤—Å–µ –∫–æ–ª–æ–Ω–∫–∏
for col in df.columns:
    analyze_column(df, col)
```

## 6Ô∏è‚É£ –¶–µ–ª–µ–≤–∞—è –ø–µ—Ä–µ–º–µ–Ω–Ω–∞—è (price)

```python
# –ü–æ–¥—Ä–æ–±–Ω—ã–π –∞–Ω–∞–ª–∏–∑ —Ü–µ–Ω—ã
print('\n' + '='*60)
print('üí∞ –ê–ù–ê–õ–ò–ó –¶–ï–õ–ï–í–û–ô –ü–ï–†–ï–ú–ï–ù–ù–û–ô: PRICE')
print('='*60)

print(f'\n–û—Å–Ω–æ–≤–Ω–∞—è —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞:')
print(f'  –°—Ä–µ–¥–Ω–µ–µ: {df["price"].mean():,.0f} ‚Ç∏')
print(f'  –ú–µ–¥–∏–∞–Ω–∞: {df["price"].median():,.0f} ‚Ç∏')
print(f'  –ú–æ–¥–∞: {df["price"].mode()[0]:,.0f} ‚Ç∏')
print(f'  –ú–∏–Ω–∏–º—É–º: {df["price"].min():,.0f} ‚Ç∏')
print(f'  –ú–∞–∫—Å–∏–º—É–º: {df["price"].max():,.0f} ‚Ç∏')
print(f'  –†–∞–∑–º–∞—Ö: {df["price"].max() - df["price"].min():,.0f} ‚Ç∏')
print(f'  –°—Ç–¥. –æ—Ç–∫–ª.: {df["price"].std():,.0f} ‚Ç∏')
print(f'  –ö–æ—ç—Ñ. –≤–∞—Ä–∏–∞—Ü–∏–∏: {(df["price"].std() / df["price"].mean()) * 100:.1f}%')

print(f'\n–ö–≤–∞—Ä—Ç–∏–ª–∏:')
print(f'  Q1 (25%): {df["price"].quantile(0.25):,.0f} ‚Ç∏')
print(f'  Q2 (50%): {df["price"].quantile(0.50):,.0f} ‚Ç∏')
print(f'  Q3 (75%): {df["price"].quantile(0.75):,.0f} ‚Ç∏')
print(f'  IQR: {df["price"].quantile(0.75) - df["price"].quantile(0.25):,.0f} ‚Ç∏')
```

```python
# –ë—ã—Å—Ç—Ä–∞—è –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è —Ü–µ–Ω—ã
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# –ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞
axes[0].hist(df['price'], bins=50, edgecolor='black', alpha=0.7)
axes[0].axvline(df['price'].mean(), color='red', linestyle='--', linewidth=2, label='–°—Ä–µ–¥–Ω–µ–µ')
axes[0].axvline(df['price'].median(), color='green', linestyle='--', linewidth=2, label='–ú–µ–¥–∏–∞–Ω–∞')
axes[0].set_xlabel('–¶–µ–Ω–∞ (‚Ç∏)', fontsize=12)
axes[0].set_ylabel('–ß–∞—Å—Ç–æ—Ç–∞', fontsize=12)
axes[0].set_title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Ü–µ–Ω', fontsize=14, fontweight='bold')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# Box plot
axes[1].boxplot(df['price'], vert=True)
axes[1].set_ylabel('–¶–µ–Ω–∞ (‚Ç∏)', fontsize=12)
axes[1].set_title('Box Plot —Ü–µ–Ω', fontsize=14, fontweight='bold')
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## 7Ô∏è‚É£ –ì–µ–æ–≥—Ä–∞—Ñ–∏—á–µ—Å–∫–∏–π –∞–Ω–∞–ª–∏–∑ (–≥–æ—Ä–æ–¥–∞)

```python
# –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –æ–±—ä—è–≤–ª–µ–Ω–∏–π –ø–æ –≥–æ—Ä–æ–¥–∞–º
print('\nüèôÔ∏è –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ –≥–æ—Ä–æ–¥–∞–º:')
city_counts = df['city'].value_counts()
print(f'\n–í—Å–µ–≥–æ –≥–æ—Ä–æ–¥–æ–≤: {df["city"].nunique()}')
print(f'\n–¢–æ–ø-10 –≥–æ—Ä–æ–¥–æ–≤ –ø–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É –æ–±—ä—è–≤–ª–µ–Ω–∏–π:')
for i, (city, count) in enumerate(city_counts.head(10).items(), 1):
    pct = (count / len(df)) * 100
    print(f'{i:2d}. {city:20s}: {count:5,} –æ–±—ä—è–≤–ª–µ–Ω–∏–π ({pct:5.1f}%)')
```

```python
# –°—Ä–µ–¥–Ω—è—è —Ü–µ–Ω–∞ –ø–æ –≥–æ—Ä–æ–¥–∞–º
print('\nüí∞ –°—Ä–µ–¥–Ω—è—è —Ü–µ–Ω–∞ –ø–æ –≥–æ—Ä–æ–¥–∞–º (—Ç–æ–ø-10):')
city_price = df.groupby('city')['price'].agg(['mean', 'median', 'count']).sort_values('mean', ascending=False)
city_price['mean'] = city_price['mean'].apply(lambda x: f'{x:,.0f} ‚Ç∏')
city_price['median'] = city_price['median'].apply(lambda x: f'{x:,.0f} ‚Ç∏')
print(city_price.head(10))
```
